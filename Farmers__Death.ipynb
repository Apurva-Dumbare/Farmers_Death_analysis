```python
total=df[['Farmers/Cultivators - 2017',
          'Farmers/Cultivators - 2018',
          'Farmers/Cultivators - 2019',
          'Farmers/Cultivators - 2020',
          'Farmers/Cultivators - 2021']].sum()
```

```python
total.plot(kind='bar', color='skyblue')
plt.title('Total Farmers by Year')
plt.xlabel('Year')
plt.ylabel('Number of Farmers')
plt.show()
```

```python
df_sorted = df.sort_values('Farmers/Cultivators - 2019', ascending=False).head(10)

plt.figure(figsize=(12, 8))
plt.bar(df_sorted['State/UT'], df_sorted['Farmers/Cultivators - 2019'], color='red')
plt.title('State-wise Farmer Deaths in 2021')
plt.xlabel('States')
plt.ylabel('Number of Farmer Deaths')
plt.xticks(rotation=90)
plt.show()
```

```python
df = df[df['Sl. No.'] != 'Total (All India)']
```

```python
df.tail(3)
```

```python
df_sorted = df.sort_values('Farmers/Cultivators - 2021', ascending=False).head(10)

# Plot horizontal bar chart
plt.barh(df_sorted['State/UT'], df_sorted['Farmers/Cultivators - 2021'], color='orange')
plt.title('Farmers by State in 2021')
plt.xlabel('Number of Farmers')
plt.ylabel('State/UT')
plt.show()
```
